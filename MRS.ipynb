```python
import pandas as pd
import numpy as np
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
import ast
# Check if 'movies' is a tuple
if isinstance(movies, tuple):
    print("'movies' is a tuple")

# Check if 'movies' is a DataFrame
elif isinstance(movies, pd.DataFrame):
    print("'movies' is a DataFrame")

# Check if 'credits' is a tuple
if isinstance(credits, tuple):
    print("'credits' is a tuple")

# Check if 'credits' is a DataFrame
elif isinstance(credits, pd.DataFrame):
    print("'credits' is a DataFrame")
    
merged_data = pd.merge(movies, credits, on='title')
movies = merged_data[['movie_id', 'title', 'overview', 'genres', 'keywords','cast', 'crew']]
print(movies)
movies.isnull().sum()
movies.dropna(inplace=True)
movies.duplicated().sum()
movies.iloc[0].genres
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def convert2(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i['name'])
            break
    return L
movies['crew'] = movies['crew'].apply(convert2)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer

# Instantiate the Porter Stemmer
ps = PorterStemmer()

# Define the stem function
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

# Apply the stem function to the 'tags' column in the DataFrame
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words= 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    # Find the index of the input movie in the DataFrame
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # Get the similarity scores for the input movie
    distances = similarity[movie_index]
    
    # Find the indices of the top 5 movies with highest similarity
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print the indices of recommended movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

# Assuming 'similarity' is the similarity matrix you have calculated
# and 'new_df' is the DataFrame containing movie information
recommend('Avatar')  # Replace 'Avatar' with the title of the movie for which you want recommendations
```
